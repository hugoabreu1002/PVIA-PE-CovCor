```python
import pandas as pd
import requests
import time
from io import StringIO
import numpy as np
from tqdm import tqdm
```

```python
def set_url(url_base, data_dic):
    url = url_base
    for key in data_dic.keys():
        url += str(key)+'='+str(data_dic[key]) + '&'
    url = url[:-1]
    return url

def get_dataframe_from_text(text):
    s=text
    substring_data = s[s.find('Month'):s.find('\r\n\tAOI loss')]
    substring_data = substring_data.replace('\t\t',',')
    substring_data = substring_data.replace('\r','')
    TESTDATA = StringIO(substring_data)
    df = pd.read_csv(TESTDATA, sep=",")
    return df

def set_df_pvgis_format(df_in, coluna_selecionada):
    values = df_in.loc[:,[coluna_selecionada]].values
    a = values.copy()
    a = np.delete(a, -1)
    a = np.insert(a, 0, values[-1])
    return a
```

```python
df_direct = pd.read_csv('radiacao_PE/direct_normal_means_PE.csv', sep=';')
df_direct.head()
```

```python
df_pvgis_sarah_Em = df_direct.copy()
df_pvgis_sarah_Hm = df_direct.copy()
df_pvgis_sarah_SDm = df_direct.copy()

df_pvgis_nsrdb_Em = df_direct.copy()
df_pvgis_nsrdb_Hm = df_direct.copy()
df_pvgis_nsrdb_SDm = df_direct.copy()
```

```python
parametros = {'lat': -8.412, 'lon': -37.592, 'raddatabase':'PVGIS-SARAH', 'peakpower':1, 'loss':14, 'inclined_optimum':1, 'vertical_optimum':1}
url_base = 'http://re.jrc.ec.europa.eu/pvgis5/PVcalc.php?'
url = set_url(url_base, parametros)
r = requests.post(url)
df = get_dataframe_from_text(r.text)
print(r.text)
df.head(13)
```

```python
parametros = {'lat': -8.412, 'lon': -37.592, 'raddatabase':'PVGIS-NSRDB', 'peakpower':1, 'loss':14, 'inclined_optimum':1, 'vertical_optimum':1}
url_base = 'http://re.jrc.ec.europa.eu/pvgis5/PVcalc.php?'
url = set_url(url_base, parametros)
r = requests.post(url)
df = get_dataframe_from_text(r.text)
print(r.text)
df.head(13)
```

```python
for i in tqdm(range(0,len(df_pvgis_sarah_Em))):
    
    lat = df_pvgis_sarah_Em.iloc[i,3]
    lon = df_pvgis_sarah_Em.iloc[i,2]
    parametros = {'lat': lat, 'lon': lon, 'raddatabase':'PVGIS-SARAH', 'peakpower':1, 'loss':14, 'inclined_optimum':1, 'vertical_optimum':1}
    url_base = 'http://re.jrc.ec.europa.eu/pvgis5/PVcalc.php?'
    url = set_url(url_base, parametros)
    r = requests.post(url)
    df_vol = get_dataframe_from_text(r.text)
    df_pvgis_sarah_Em.iloc[i,4:] = set_df_pvgis_format(df_vol, 'Em')
    df_pvgis_sarah_Hm.iloc[i,4:] = set_df_pvgis_format(df_vol, 'Hm')
    df_pvgis_sarah_SDm.iloc[i,4:] = set_df_pvgis_format(df_vol, 'SDm')
    
    time.sleep(0.01)

    parametros = {'lat': lat, 'lon': lon, 'raddatabase':'PVGIS-NSRDB', 'peakpower':1, 'loss':14, 'inclined_optimum':1, 'vertical_optimum':1}
    url_base = 'http://re.jrc.ec.europa.eu/pvgis5/PVcalc.php?'
    url = set_url(url_base, parametros)
    r = requests.post(url)
    df_vol = get_dataframe_from_text(r.text)  
    df_pvgis_nsrdb_Em.iloc[i,4:] = set_df_pvgis_format(df_vol, 'Em')
    df_pvgis_nsrdb_Hm.iloc[i,4:] = set_df_pvgis_format(df_vol, 'Hm')
    df_pvgis_nsrdb_SDm.iloc[i,4:] = set_df_pvgis_format(df_vol, 'SDm')
    
    time.sleep(0.01)
```

```python
df_pvgis_sarah_Em.to_csv('PVGIS_PE\df_pvgis_sarah_Em.csv', index=None, sep=';')
df_pvgis_sarah_Hm.to_csv('PVGIS_PE\df_pvgis_sarah_Hm.csv', index=None, sep=';')
df_pvgis_sarah_SDm.to_csv('PVGIS_PE\df_pvgis_sarah_SDm.csv', index=None, sep=';')
df_pvgis_nsrdb_Em.to_csv('PVGIS_PE\df_pvgis_nsrdb_Em.csv', index=None, sep=';')
df_pvgis_nsrdb_Hm.to_csv('PVGIS_PE\df_pvgis_nsrdb_Hm.csv', index=None, sep=';')
df_pvgis_nsrdb_SDm.to_csv('PVGIS_PE\df_pvgis_nsrdb_SDm.csv', index=None, sep=';')
```
